```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import random
```

```python
#Creating data
x = [3.84,4.34,3.84,5,5.5,6,6.5,5.83]
y = [5.0,10.0,15.0,20.0,25.0,30.0,35.0,40.0]
x = np.asarray(x)
plt.title('Raw Data')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.plot(x,y,'ro')
plt.show()
```

```python
#Creating Model
m = tf.Variable(0.0) #Slope
c = tf.Variable(0.0) #y-intercept
train_y = m * x + c
loss = tf.reduce_mean(tf.square(y - train_y)) #Loss function
optimizer = tf.train.GradientDescentOptimizer(0.03) 
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```python
#Training Model
loss_values = []
for step in range(1000):
    _, loss_val, m_val, c_val = sess.run([train, loss, m, c])
    loss_values.append(loss_val)
    if step % 5 == 0:
        print(step, loss_val, m_val, c_val)
```

```python
#plotting losses
plt.title('Loss value')
plt.xlabel('Iteration Number')
plt.ylabel('Loss')
plt.plot(loss_values, 'ro')
plt.show()
#plotting line
plt.title('Data with approximation line')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.plot(x,y,'ro')
plt.plot(x,m_val * x + c_val)
plt.show()
```
